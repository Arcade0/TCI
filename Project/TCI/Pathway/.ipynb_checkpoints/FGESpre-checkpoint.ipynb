```r
S_Am = read.csv('S_Am/Total/Input/S_Am.csv',row.names = 1)
S_Dm = read.csv('Pre/S_Dm.csv',row.names = 1)
A_D = read.csv('Pre/A_D.csv',row.names = 1)
```

```r
TAD = read.csv('Pre/TAD.csv')
TAD0 = read.csv('Pre/TAD0.csv')
```

```r
####label and save block SGA and DEG####
selDEG = function(S_Ablock,TAD,TAD0, S_D, name){
    SGA = factor(colnames(S_Ablock),levels=levels(TAD[,1]))
    imputen = TAD[TAD[,1]%in%SGA,]
    SGAlist = unique(imputen[,1])
    DEGlist = unique(imputen[,2])
    S_Dblock = S_Dm[rownames(S_Ablock),DEGlist]

    SGAlist = as.character(SGAlist)
    for (i in (1:length(SGAlist))){
        SGAlist[i] = paste('sga', SGAlist[i], sep=':' )
    } 
    SGAlist = as.factor(SGAlist)

    DEGlist = as.character(DEGlist)
    for (i in (1:length(DEGlist))){
        DEGlist[i] = paste('deg', DEGlist[i], sep=':')
    }
    DEGlist = as.factor(DEGlist) 
    colnames(S_Ablock) = SGAlist
    colnames(S_Dblock) = DEGlist
    SGAlistn = paste(name, 'SGAlist.csv', sep='_')
    DEGlistn = paste(name, 'DEGlist.csv', sep='_')
    write.csv(SGAlist, SGAlistn, quote=F, row.names=F)
    write.csv(DEGlist, DEGlistn, quote=F, row.names=F)
    ####combine S_A with S_D####
    S_ADblock = cbind(S_Ablock, S_Dblock)
    S_ADblockn = paste(name, 'S_ADblock.csv', sep='_')
    write.csv(S_ADblock, S_ADblockn, quote=F, row.names=F)
    ####add forbidden knowledge####
    imputef = TAD0[TAD0[,1]%in%SGAlist,1:2]
    imputef = imputef[imputef[,2]%in%DEGlist,1:2]
    imputefn = paste(name, 'imputef.csv', sep='_')
    write.csv(imputef, imputefn, sep=' ', quote=F, row.names=F)
    print(dim(imputef))
    return(list(S_Ablock,S_Dblock))
}
```

```r
####paper pathway
TGFlist = c('TGFBR1','TGFBR2','SMAD2','SMAD3','SMAD4')
PI3Klist = c('PTEN','PIK3R2','PIK3R1','PIK3R2','PIK3CA','PIK3CB','INPP4B','AKT1','AKT2','AKT3','TSC1','TSC2','RHEB','STK11','MTOR','RICTOR','RPTOR')
RB1list = c('CDKN1A','CDKN1B','CDKN2A','CDKN2B','CDKN2C','CCNE','CCND1','CCND2','CCND3','CDK2','CDK4','CDK6','RB1','E2F1','E2F3')
TP53list = c('MDM2','MDM4','CDKN2A','TP53','ATM','CHEK2','RPS6KA3','CDKN1A','CDKN1B','CDKN2A','CDKN2B','CDKN2C','CCNE','CCND1','CCND2','CCND3','CDK2','CDK4','CDK6','RB1','E2F1','E2F3')
WNTlist = c('WIF1','SFRP1','WNT','LRP5','LRP6','DKK4','AXIN1','AXIN2','CTNNB1','GSK3B','APC','RNF43','ZNRF3','FZD','TCF7','TCF7L1','TCF7L2','TLE4')
HIPPOlist = c('DCHS1','DCHS2','FAT1','FAT2','FAT3','FAT4','TAOK1','TAOK2','TAOK3','SAV1','STK3','STK4','LATS1','LATS2','MOB1A','MOB1B','NF2','WWC1','CRB1','CRB2','YAP1','TAZ','PTPN14','CSNK1E','CSNK1D','TEAD2')
NOTCHlist = c('NOV','CNTN6','JAG2','ARRDC1','NOTCH1','NOTCH2','NOTCH3','NOTCH4','DNER','PSEN2','FBXW7','CUL1','MAML3','KAT2B','CREBBP','EP300','NCOR1','NCOR2','SPEN','KDM5A','HES.X','HEY.X')
Nrf2list = c('KEAP1','CUL3','NFE2L2')
RTKlist = c('EGFR','ERBB2','ERBB4','MET','PDGFRA','FGFR1','FGFR2','FGFR3','FGFR4','KIT','IGF1R','RET','ROS1','ALK','FLT3','NTRK1-3','JAK2','CBL','ERRFI1','ABL1','SOS1','NF1','RASA1','PTPN11','KRAS','HRAS','NRAS','RIT1','ARAF','BRAF','RAF1','RAC1','MAPK1','MAP2K1','MAP2K2')
```

```r
pathway = list(TGFlist, PI3Klist, RB1list, TP53list, WNTlist, HIPPOlist, NOTCHlist, Nrf2list, RTKlist)
pathwayname = c('Pathway/TGF/TGF','Pathway/PI3K/PI3K','Pathway/RB1/RB1','Pathway/TP53/TP53','Pathway/WNT/WNT','Pathway/HIPPO/HIPPO','Pathway/NOTCH/NOTCH','Pathway/Nrf2/Nrf2','Pathway/RTK/RTK')
```

```r
for (i in 1:length(pathway)){
    aname = paste(pathwayname[i], '0',sep='_')
    bname = paste(pathwayname[i], '1',sep='_')
    anamen = paste(aname, 'csv', sep='.')
    bnamen = paste(bname, 'csv', sep='.')
    a = S_Am[,colnames(S_Am)%in%pathway[[i]]]
    b = S_Areb[,colnames(S_Areb)%in%pathway[[i]]]
    selDEG(a,TAD,TAD0, S_D, aname)
    selDEG(b,TAD,TAD0, S_D, bname)
    write.csv(a, anamen, quote=F, row.names=T)
    write.csv(b, bnamen, quote=F, row.names=T)
}
```
